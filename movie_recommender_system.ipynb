```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

sns.set_theme()
```

```python
df = pd.read_csv("sample_data/movies.csv")
df.head()
```

```python
# extract key columns
df = df[['title', 'genres', 'keywords', 'cast', 'director']]
df = df.fillna('')
df.head()
```

```python
# combine key columns into one column called Features
df['features'] = df['title'] + ' ' + df['genres'] + ' ' + df['keywords'] + ' ' + df['cast'] + ' ' + df['director']
```

```python
# vectorize
vectorizer = CountVectorizer(stop_words="english", min_df=20)
word_matrix = vectorizer.fit_transform(df["features"])
word_matrix.shape
```

```python
# compute similarity
sim = cosine_similarity(word_matrix)
```

```python
sim.shape
```

```python
list(enumerate(sim[0]))
```

```python
def get_recommendations(title, df, sim, count=10):
    # title index
    index = df.index[df["title"].str.lower() == title.lower()]

    # no entry for specified title
    if len(index) == 0:
        return []
    
    # get corresponging row in similarity matrix
    similarities = list(enumerate(sim[index[0]]))

    # sort in descending order
    recommendations = sorted(similarities, key=lambda x : x[0], reverse=True)

    # get top n, ignore first entry since it represents the title itself with a similarity of 1.0
    top_recs = recommendations[1:count+1]

    # generate list of titles from indices
    titles = []

    for i in range(len(top_recs)):
        title = df.iloc[top_recs[i][0]]["title"]
        titles.append(title)

    return titles    
```

```python
# search for similar movies
get_recommendations('Skyfall', df, sim)
```

```python
get_recommendations('Mulan', df, sim)
```
